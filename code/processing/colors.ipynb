### Creamos vector de colores

#### Obtenemos todos los colores posibles

```python
def process_color_str(text):
    if text is not np.nan:
        colors_processed = text.replace('y', ' ').replace(',', ' ')
        colors_splitted = colors_processed.split(' ')

    colors = list()
    for raw_color in colors_splitted:
        if len(raw_color) > 0 and raw_color != ',':
            colors.append(raw_color)
    return colors
```

```python
def process_color_combination(df, col):
    df[col] = df[col].str.lower()

    colors_combs = df[col].unique().tolist()
    colors_combs_processed = [col.replace('y', ' ').replace(',', ' ') for col in colors_combs if col is not np.nan]
    colors_combs_splitted = [col.split(' ') for col in colors_combs_processed]

    final_combs = list()
    for comb in colors_combs_splitted:
        new_comb = list()
        for col in comb:
            if len(col) > 0 and col != ',':
                new_comb.append(col)
        new_comb.sort()
        final_combs.append(new_comb)

    colors = list()
    for comb in final_combs:
        colors += comb

    colors = list(set(colors))

    return colors

raw_colors = process_color_combination(df, 'CUERPO_seccion1_EXTERNO_Color')
raw_colors
```

```python
colors = set()
for c in raw_colors:
    if c in color_mapper:
        colors.add(color_mapper[c])
    else:
        colors.add(c)

colors = sorted(list(colors))
colors
```

```python
color_columns = ['CUERPO_seccion1_EXTERNO_Color', 'CUERPO_seccion1_INTERNO_Color']
```

```python
def process_color_str(text, color, mapper):
    text = text.lower()
    colors_processed = text.replace('y', ' ').replace(',', ' ')
    colors_splitted = colors_processed.split(' ')

    raw_colors = list()
    for raw_color in colors_splitted:
        if len(raw_color) > 0 and raw_color != ',':
            raw_colors.append(raw_color)

    colors = [mapper[col] for col in raw_colors]
    if color in colors:
        return True
    return False
```

```python
def check_if_color_in_columns(row, color, color_mapper):
    cols = row.index.tolist()

    for col in cols:
        raw_color_text = row[col]
        if raw_color_text is not np.nan:
            color_in_column = process_color_str(raw_color_text, color, color_mapper)
            if color_in_column:
                return 1
        else:
            continue
    return 0

```

```python
for c in colors:
    df[f'trait_color_{c}'] = df[color_columns].apply(lambda x: check_if_color_in_columns(x, c, color_mapper), axis=1)
```
