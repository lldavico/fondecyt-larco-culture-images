```python
def handle_not_spaces(row):
    if row is not np.nan:
        uppercase_indexes = [i for i in range(len(row)) if row[i].isupper()][1:]
        new_row = list(row)
        for j in uppercase_indexes:
            j = j + 1*(uppercase_indexes.index(j))
            new_row.insert(j, ' ')

        row = ''.join(new_row)    
    return row

handle_not_spaces('PintadoEscultóricoRelieveInciso')
```

### Manejo tildes

```python
def handle_accent_mark(value):
    '''
    If string value is not NaN, returns a lowered string without
    accent marks
    '''

    normalization = {
        'á': 'a',
        'é': 'e',
        'í': 'i',
        'ó': 'o',
        'ú': 'u'
    }

    if value is not np.nan:
        accented_chars = list(normalization.keys())

        new_value = ''
        for c in value.lower():
            if c in accented_chars:
                new_value += normalization[c]
            else:
                new_value += c

        return new_value 
    
    return value
```

```python
df['Categoría Morfofuncional'] = df['Categoría Morfofuncional'].apply(handle_accent_mark)
df['Descripción'] = df['Descripción'].apply(handle_accent_mark)
df['Escena Principal'] = df['Escena Principal'].apply(handle_accent_mark)
df['CUERPO_seccion1_EXTERNO_Técnica Decoración'] = df['CUERPO_seccion1_EXTERNO_Técnica Decoración'].apply(handle_accent_mark)
df['CUERPO_seccion1_EXTERNO_Color'] = df['CUERPO_seccion1_EXTERNO_Color'].apply(handle_accent_mark)
df['CUERPO_seccion1_INTERNO_Color'] = df['CUERPO_seccion1_INTERNO_Color'].apply(handle_accent_mark)
df['Acabado'] = df['Acabado'].apply(handle_accent_mark)

df
```

#### Tipo condición: 'menciona'

```python
def get_mention_word(añadir_str):
    if 'escultoric' in añadir_str:
        if 'animal' in añadir_str:
            return [('escultoric', 'escultura'), ('animal',)]  
        return [('escultoric', 'escultura')]
    elif 'tripode' in añadir_str:
        return [('tripode',)]
    elif 'pedestal' in añadir_str:
        return [('pedestal',)]
    elif 'pintad' in añadir_str:
        if 'geometric' in añadir_str:
            return [('pintad', 'pintura'), ('geometric',)]    
        elif 'animal' in añadir_str:
            return [('pintad', 'pintura'), ('animal',)]  
        return [('pintad', 'pintura')]
    

def add_mention_words(row):
    if row['añadir'] is not np.nan:
        añadir_str = row['añadir'].lower()
        if añadir_str.count('menciona') > 0 and 'no menciona' not in añadir_str:
            mentioned_words = get_mention_word(añadir_str)
            return mentioned_words
        return np.nan
    else:
        return np.nan
```

```python
cat_df_filtered['menciona'] = cat_df_filtered.apply(lambda x: add_mention_words(x), axis=1)
```

#### Tipo condicion: 'No menciona'

```python
def add_non_mention_words(row):
    if row['añadir'] is not np.nan:
        añadir_str = row['añadir'].lower()
        if 'no menciona "base"' in añadir_str:
            return ('base',)
        elif 'no menciona "asa"' in añadir_str:
            return ('asa',)
        return np.nan
    else:
        return np.nan
```

```python
cat_df_filtered['no_menciona'] = cat_df_filtered.apply(lambda x: add_non_mention_words(x), axis=1)
```

#### Tipo condicion: 'trait'

```python
def get_trait_list(añadir_str):
    trait_str = añadir_str.split(' ')[0]
    trait_list = trait_str.replace('trait_n=', '').split('-')
    return trait_list


def is_trait(row):
    if row['añadir'] is not np.nan:
        añadir_str = row['añadir'].lower()
        if añadir_str.count('trait') > 0 and 'no es trait' not in añadir_str:
            traits = get_trait_list(añadir_str)
            return traits
        return np.nan
    else:
        return np.nan
```

```python
cat_df_filtered['traits'] = cat_df_filtered.apply(lambda x: is_trait(x), axis=1)
```

#### Tipo condicion: 'No trait'

```python
def get_not_trait_list(añadir_str):
    trait_str = añadir_str.split('trait_n=')[1].split(' ')[0]
    trait_list = trait_str.split('-')
    return trait_list


def is_not_trait(row):
    if row['añadir'] is not np.nan:
        añadir_str = row['añadir'].lower()
        if 'no es trait' in añadir_str:
            traits = get_not_trait_list(añadir_str)
            return traits
        return np.nan
    else:
        return np.nan
```

```python
cat_df_filtered['no_traits'] = cat_df_filtered.apply(lambda x: is_not_trait(x), axis=1)
```

### Calculamos la frecuencia de aparicion de cada palabra de la lista en las columnas

#### Función para chequear que palabra esté en el texto. Acá se debe agregar la condición descrita en 'añadir'

```python
# Reemplazar esta funcion por lógica más inteligente: embeddings
def word_in_text(words, text):
    if isinstance(words, str):
        return True if words in text else False
    elif isinstance(words, tuple) or isinstance(words, list):
        for w in words:
            if w in text:
                return True
        return False
    
def word_not_in_text(words, text):
    if isinstance(words, str):
        return True if not words in text else False
    elif isinstance(words, tuple) or isinstance(words, list):
        for w in words:
            if w in text:
                return False
        return True
    
def trait_appear(trait_row, ids):
    trait_ids = [f'trait_n{i}' for i in range(int(ids[0]), int(ids[1])+1)]
    trait_appears = list(trait_row[trait_ids].values[0])
    if any(v > 0 for v in trait_appears):
        return True
    return False

        
def trait_not_appear(trait_row, ids):
    trait_ids = [f'trait_n{i}' for i in range(int(ids[0]), int(ids[1])+1)]
    trait_appears = list(trait_row[trait_ids].values[0])
    if all(v == 0 for v in trait_appears):
        return True
    return False
    
```

```python
def get_words_freq_in_columns(row, word_conditions, freq_df_before):

    columns = row.index.values.tolist()
    _id = row['catalogation_id']
    words_appearances = 0

    text_in_columns = [row[col] for col in columns[1:] if row[col] is not np.nan]
    text = ' | '.join(text_in_columns)

    # Condición 'trait'
    if word_conditions['traits'] is not np.nan:
        id_trait_row = freq_df_before[freq_df_before['catalogation_id'] == _id]
        trait_result = trait_appear(id_trait_row, word_conditions['traits'])
        if not trait_result:
            return 0

    # Condición 'no trait'
    if word_conditions['no_traits'] is not np.nan:
        id_trait_row = freq_df_before[freq_df_before['catalogation_id'] == _id]
        no_trait_result = trait_not_appear(id_trait_row, word_conditions['no_traits'])
        if not no_trait_result:
            return 0
    
    # Condición 'words'
    words_in_condition = False
    if len(word_conditions['words']) > 0:
        words_in_condition = True
        for word in word_conditions['words']:
            if word_in_text(word, text):
                words_appearances += 1
        if words_appearances == 0:
            return 0
        
    # Condición 'mention'
    if word_conditions['mention'] is not np.nan:
        mentioned_results = [word_in_text(comb, text) for comb in word_conditions['mention']] 
        if False in mentioned_results:
            return 0
        
    # Condición 'non mention'
    if word_conditions['non_mention'] is not np.nan:
        non_mentioned_results = [word_not_in_text(w, text) for w in word_conditions['non_mention']]
        if False in non_mentioned_results:
            return 0

    # Evaluacion de las condiciones
    if words_in_condition:
        total_freq = words_appearances
    else:
        total_freq = 1

    return total_freq
```
